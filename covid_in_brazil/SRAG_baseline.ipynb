```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
df_train.shape
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
X_train = df_train.select_dtypes(['number']).fillna(0).drop(columns='EVOLUCAO')
y_train = df_train['EVOLUCAO']

X_test = df_test.select_dtypes(['number']).fillna(0)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.10, random_state=42)
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
acc = clf.score(X_val, y_val)


print(f"{acc*100:.2f}%")
```

```python
val_pred = clf.predict_proba(X_val)[:,1]
```

```python
from sklearn.metrics import f1_score
f1_score(y_val, np.round(val_pred))
```

```python
test_pred = clf.predict_proba(X_test)[:,1]
```

```python
test_pred
```

```python
pd.DataFrame(np.round(test_pred)).reset_index().to_csv('out_lr.csv', header=['ID', 'EVOLUCAO'], index=False)
```
